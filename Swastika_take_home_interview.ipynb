```python
m1["timestamp_utc"]=pd.to_datetime(m1["timestamp_utc"])
def convert_timezone(row):
    dt_utc=row['timestamp_utc']
    timezone=row['timezone_str']
    tz=pytz.timezone(timezone)
    timestamp_utc=dt_utc.astimezone(tz)
    return timestamp_utc

m1["req_time"]=m1.apply(convert_timezone, axis=1)
m1["req_time"] = m1["req_time"].apply(lambda x: x.strftime('%H:%M:%S'))
```

```python
m=pd.merge(m1,buss_time, on="store_id")
m
```

```python
import pandas as pd

# Load the data
data = pd.read_csv('C:/Users/swast/Desktop/fil_output1.csv')

data['timestamp_utc'] = pd.to_datetime(data['timestamp_utc'])

# Extract date component and assign to new column
data['date'] = data['timestamp_utc'].dt.date

# Convert date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter for successful requests
success_data = data[data['status'] == 'active']

# Convert start and end time columns to datetime type
success_data['start_time_local'] = pd.to_datetime(success_data['start_time_local'])
success_data['end_time_local'] = pd.to_datetime(success_data['end_time_local'])

# Calculate the uptime_last_day
success_data['uptime_last_day'] = (success_data['end_time_local'] - success_data['start_time_local']).dt.total_seconds() / 3600

# Calculate the uptime_last_week
success_data['uptime_last_week'] = success_data.groupby(['store_id', pd.Grouper(key='timestamp_utc', freq='W-MON')])['uptime_last_day'].transform('sum')

# Group the data by store_id and date and aggregate the uptime_last_day and uptime_last_week
uptime = success_data.groupby(['store_id', 'date'])['uptime_last_day', 'uptime_last_week'].sum().reset_index()

# Filter for inactive requests
inactive_data = data[data['status'] == 'inactive']

# Convert start and end time columns to datetime type
inactive_data['start_time_local'] = pd.to_datetime(inactive_data['start_time_local'])
inactive_data['end_time_local'] = pd.to_datetime(inactive_data['end_time_local'])

# Calculate the downtime_last_day
inactive_data['downtime_last_day'] = (inactive_data['end_time_local'] - inactive_data['start_time_local']).dt.total_seconds() / 3600

# Calculate the downtime_last_week
inactive_data['downtime_last_week'] = inactive_data.groupby(['store_id', pd.Grouper(key='timestamp_utc', freq='W-MON')])['downtime_last_day'].transform('sum')

# Group the data by store_id and date and aggregate the downtime_last_day and downtime_last_week
downtime = inactive_data.groupby(['store_id', 'date'])['downtime_last_day', 'downtime_last_week'].sum().reset_index()

# Merge the two dataframes on 'store_id' and 'date'
combined_data = pd.merge(uptime, downtime, on=['store_id', 'date'], how='outer')

# Fill any missing values with 0
combined_data1 = combined_data.fillna(0)

print(combined_data1)
```
